## Word2Vec

```python
import gensim
```

```python
from gensim.models import Word2Vec, KeyedVectors
```

```python
import gensim.downloader as api

wv = api.load('word2vec-google-news-300') # wv - Word to vec wv
```

```python
vec_king = wv['king']
vec_king
```

```python
wv.most_similar('cricket')
```

```python
wv.similarity('hockey', 'cricket')
```

```python
wv.similarity('hockey', 'sports')
```

```python
vec = wv['king'] - wv['man'] + wv['queen']
vec
```

```python
wv.most_similar(vec)
```
